```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import opendatasets as od
od.download("https://www.kaggle.com/datasets/hemendrasr/pizza-vs-ice-cream")
```

```python
import os
directory = '/content/pizza-vs-ice-cream/dataset'
contents = os.listdir(directory)
print(contents)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data='/content/pizza-vs-ice-cream/dataset/train'
test_data='/content/pizza-vs-ice-cream/dataset/test'
```

```python
train_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_generator=train_datagen.flow_from_directory(
train_data,
target_size=(256,256),
    batch_size=32,
    class_mode='binary'
)
test_generator=test_datagen.flow_from_directory(
test_data,
target_size=(256,256),
    batch_size=32,
    class_mode='binary'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
model = Sequential()
model.add(Flatten(input_shape=(256, 256, 3)))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
history=model.fit(train_generator,epochs=100,validation_data=test_generator)
```

```python
train_accuracy = history.history['accuracy'][-1]
print(train_accuracy)
```

```python
y_prob=model.predict(test_generator)
y_pred=y_prob.argmax(axis=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
